<h1 style="font-size:20; text-align:center; color:red; font-family:bold">Importation des données</h1>

```python
# Importez la librairie Pandas

import pandas as pd
```

```python
# Sources des données

file = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'

# Importez les données en tant que dataframe pandas

df = pd.read_csv(file)

# Affichez les 10 premières lignes de la dataframe

df.head(10)
```

Les données n'ont pas une ligne de noms de colonnes. Vous remarquez que la première observationn a été utilisée comme noms de colonnes. Pour corrisger cela, vous devez retrouvez les noms de colonnes à partir de ce lien https://archive.ics.uci.edu/ml/datasets/automobile et les stocker dans une liste

```python
# Créez une liste de stockage des noms de colonnes

colones = ["symbling", "normalized-losses", "make",
            "fuel-type", "aspiration", "num-of-doors",
            "body-style", "drive-wheels", "engine-location",
            "wheel-base", "length", "width", "height",
            "curb-weight", "engine-type", "num-of-cylinders",
            "engine-size", "fuel-system", "bore", "stroke",
            "compression-ratio", "horsepower", "paek-rmp",
            "city-mpg", "highway-mpg", "price"]

# Afficher la liste : colonnes

colonnes
```

```python
# Réimportez les données tout en spécifiant les noms des colonnes

df = pd.read_csv(file, header = None, names = colones)

# Affichez les premières cinq lignes de la dataframe

df.head()

```

```python
# Afficher les dernières lignes
df.tail(10)
```

Pour finir cette première partie concernar l'importation des données en tant que dataframe, vous devez sauvegarder la dataframe au format csv afin de pouvoir l'utiliser ultérieurement. Pour réaliser cette tâche de sauvegarde, vous utiliserez la méthode <strong>to_csv()</strong>

```python
# Sauvegarder de la dataframe au format csv

df.to_csv('automibile.csv', index = False)
```

Le code ci-dessus m'a permis de sauvegarder la dataframe sous le nom automobile.csv dans un dossier précis de mon Drive. Ici, l'argument <strong>index</strong> a pour <strong>False</strong> car la dataframe contient déjà des indices lignes et on ne vaut pas en créer d'autres.
